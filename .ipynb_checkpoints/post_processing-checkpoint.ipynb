```python
#점과 점사이 유클리드 거리 함수 정의
import math
def distance(x1, y1, x2, y2):
    result = math.sqrt( math.pow(x1 - x2, 2) + math.pow(y1 - y2, 2))
    return result
import os
import cv2
import json
import numpy as np
import pandas as pd
from moviepy.editor import *
import matplotlib.pyplot as plt
import shutil
```

interpolation! --> 추가 수정 필요! (interpolration 안된 부분 존재)

```python
PATH = "C:/Users/iai/Desktop/video_extraction/outputs/001"
#25까지 완료
for action in os.listdir(PATH):
    JSON = PATH + "/" + action+ "/"+action+".json"
    with open(JSON, encoding = 'utf-8') as json_file:
        data = json.load(json_file)
    #np.nan 보간
    data_list = []
    for coord in data:
        data_dict = {}
        for ball in coord.keys():
            if type(coord[ball]) == float:
                data_dict[ball+"_x"] = np.nan
                data_dict[ball+"_y"] = np.nan
            else:
                data_dict[ball+"_x"] = coord[ball][0]
                data_dict[ball+"_y"] = coord[ball][1]

        data_list.append(data_dict)
    #모든 coord에 대해 변환 완료
    df = pd.DataFrame(data_list)
    df1 = df.interpolate()
    #보간 완료
    inter_list = df1.to_dict('records')
    #list 변환 완료
    final_list = []
    for coord in inter_list:
        coord_dict = {}
        coord_dict["white ball"] = [coord['white ball_x'],coord['white ball_y']]
        coord_dict["red ball"] = [coord['red ball_x'],coord['red ball_y']]
        coord_dict["yellow ball"] = [coord['yellow ball_x'],coord['yellow ball_y']]
        final_list.append(coord_dict)
    #final_list에 보간 완료한 좌표 데이터 저장

    #마지막에 저장!
    DATA = PATH + "/" + action+ "/data.json"
    with open(DATA, 'w', encoding="utf-8") as outfile:
        json.dump(final_list, outfile,  ensure_ascii=False, indent="\t")
    #폴더에 data.json 파일 저장
#1~26까지 완료!
```

```python
PATH = "C:/Users/iai/Desktop/video_extraction/outputs/026"
#25까지 완료!
for action in os.listdir(PATH):
    #하나의 비디오에 대해
    MP4 = PATH + "/" + action +"/"+ action+".mp4"
    DATA = PATH + "/" + action+ "/data.json"
    cap = cv2.VideoCapture(MP4)
    length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    time = int(length/fps)
    #하나의 action 안에서
    if time >= 40:
        print(action, "동영상 중복 검사 시작")
        # 2개로 자르기 
        with open(DATA, encoding = 'utf-8') as json_file:
            final_list = json.load(json_file)

        #복사하기
        final_list1 = final_list.copy()
        #final_list에 보간 완료한 좌표 데이터 저장      
        frame_count = []
        json_list =[]
        json_count = 0
        start = 0
        
        
        for i in range(500, len(final_list)-103):            
            if (((distance(final_list[i]["white ball"][0],final_list[i]["white ball"][1], final_list[i+100]["white ball"][0], final_list[i+100]["white ball"][1] ) < 1)
            and (distance(final_list[i]["red ball"][0],final_list[i]["red ball"][1], final_list[i+100]["red ball"][0], final_list[i+100]["red ball"][1] ) < 1)
            and (distance(final_list[i]["yellow ball"][0],final_list[i]["yellow ball"][1], final_list[i+100]["yellow ball"][0], final_list[i+100]["yellow ball"][1] ) < 1))
            and ((distance(final_list[i+100]["white ball"][0],final_list[i+100]["white ball"][1], final_list[i+101]["white ball"][0], final_list[i+101]["white ball"][1] ) >= 3 )   
            or (distance(final_list[i+100]["red ball"][0],final_list[i+100]["red ball"][1], final_list[i+101]["red ball"][0], final_list[i+101]["red ball"][1] ) >= 3 )  
            or (distance(final_list[i+100]["yellow ball"][0],final_list[i+100]["yellow ball"][1], final_list[i+101]["yellow ball"][0], final_list[i+101]["yellow ball"][1] ) >= 3))) :
                
                #i+156부터 움직이기 시작
                #앞부분부터 전부 넣기
                #멈춰있으면 frame_count에 해당 프레임 번호 넣기
                json_list.append(final_list[start:i+30])
                    
                try:
                    if not os.path.exists(PATH + "/"+action+"_{:01d}".format(json_count)):
                        os.makedirs(PATH + "/"+action+"_{:01d}".format(json_count))

                except OSError:
                    print ('Error: Creating directory. ' +  PATH + "/"+action+"_{:01d}".format(json_count))
                
                json_name = PATH + "/"+action+"_{:01d}/".format(json_count)+"data.json"
                with open(json_name, 'w', encoding="utf-8") as outfile:
                    json.dump(json_list[0], outfile,  ensure_ascii=False, indent=4)
                    #폴더에 json 파일 추가로 저장
                print(action, "/json_count = ",json_count , "앞부분 분리 완료")
                #복사해둔 곳에서 삭제
                del final_list1[:i+30]
                frame_count.append(i+30)
                json_count += 1
                start = i+30
                json_list.clear()
        
        #앞부분 분리 완료
        try:
            if not os.path.exists(PATH + "/"+action+"_{:01d}".format(json_count)):
                os.makedirs(PATH + "/"+action+"_{:01d}".format(json_count))

        except OSError:
            print ('Error: Creating directory. ' +  PATH + "/"+action+"_{:01d}".format(json_count))

        json_name = PATH + "/"+action+"_{:01d}/".format(json_count)+"data.json"
        with open(json_name, 'w', encoding="utf-8") as outfile:
            json.dump(final_list1, outfile,  ensure_ascii=False, indent=4)
            #폴더에 json 파일 추가로 저장
        print(action, "/json_count = ",json_count , "마지막 분리 완료")    
        
                
        #우선 한 데이터 전부다 확인

        #이 while문 안에서는 한프레임에 해당
        save_img_path= []
        video_count = 0
        count = 0
        
        
        cap1 = cv2.VideoCapture(MP4)    
        while (cap1.isOpened()):
            ret, frame = cap1.read()
                
            #frame_count안에 뭐라도 있다면 save_img_path를 가지고 동영상 만들기
            if (len(frame_count)!=0):
                if(count == frame_count[0]):
                    print(action, "/video_count = ",video_count, "동영상 분리 완료")
                    clip = ImageSequenceClip([cv2.cvtColor(frame, cv2.COLOR_RGB2BGR) for frame in save_img_path], fps=int(fps))
                    result_video_name = PATH + "/"+action+"_{:01d}/".format(video_count)+action+"_{:01d}.mp4".format(video_count)
                    clip.write_videofile(result_video_name)
                    save_img_path.clear()
                    video_count += 1
                    frame_count.pop(0)     
            # 동영상 전부 복사됐다면 frame_count가 없었다는 의미!
            if (np.shape(frame) == () and len(save_img_path) != 0 ):
                print(action, "/json_count = ",json_count,"마지막 동영상 분리 완료")
                result_video_name = PATH + "/"+action+"_{:01d}/".format(video_count)+action+"_{:01d}.mp4".format(video_count)
                del save_img_path[-1]
                clip = ImageSequenceClip([cv2.cvtColor(frame, cv2.COLOR_RGB2BGR) for frame in save_img_path], fps=int(fps))
                clip.write_videofile(result_video_name)
                break;
                
            save_img_path.append(frame)
            count += 1 
            
        cap1.release()
    print(action, "동영상 중복 검사 완료")    
    cap.release()
print("모든 동영상 완료!")
#26까지 모든 동영상 완료
```

```python
#공 지정
PATH = "C:/Users/iai/Desktop/video_extraction/outputs/026"

for action in os.listdir(PATH):
    #하나의 비디오에 대해
    MP4 = PATH + "/" + action +"/"+ action+".mp4"
    DATA = PATH + "/" + action+ "/data.json"

    #json 파일 경로
    with open(DATA, encoding = 'utf-8') as json_file:
        final_list = json.load(json_file)
    #리스트 형식으로 받아옴

    # cue ball, 1_object_ball, 2_object_ball 저장 
    ball_list = ["white ball", "red ball", "yellow ball"]
    info_dict = {'white ball': "?", 'yellow ball': "?", 'red ball': "?", 'skill': "?"}

    for i in range(1,len(final_list)-150):
        if (len(ball_list)==3):
            for ball in ball_list:

                if (distance(final_list[i][ball][0],final_list[i][ball][1], final_list[i-1][ball][0], final_list[i-1][ball][1] ) > 2):
                    #움직이면 바로 그 공 이름 지정하고 공 제외시키기.
                    info_dict[ball] = "cue_ball"
                    ball_list.remove(ball)
                    #만약 3번째의 white ball이 걸리면 그대로 빠져나가기
                    break;
            #빠져나와서 여기가 된다.
            #두개가 아니면 해당하지 않는다. (3번째의 red ball 부터)
        if (len(ball_list) == 2):
            #남은 두개 가지고 먼저 움직이는것 1목적구 지정
            for ball in ball_list:
                if (distance(final_list[i][ball][0],final_list[i][ball][1], final_list[i-1][ball][0], final_list[i-1][ball][1] ) > 2):
                    info_dict[ball] = "1_object_ball"
                    ball_list.remove(ball)
                    break;
        if (len(ball_list) ==1) :
            info_dict[ball_list[0]]= "2_object_ball"
            break;
            #i 전부 끝내기!
            #final_list 에 값담기 완료
            
    print(action, info_dict)



        
    #움직이면 바로 지정하고 공 제외시키기
    #수구는 확실!
    INFO =  PATH + "/" + action +"/info.json"

    with open(INFO, 'w', encoding="utf-8") as outfile:
        json.dump(info_dict, outfile,  ensure_ascii=False, indent=4)
    #폴더에 json 파일 저장
            
        
        
#000만 final 가지고 있음!





        #rule-base 어려움 : 아예 안맞추는 경우도 발생!! 

       # cue ball 다음 
#         좌표 앞부분 자르기  
#         공 세개의 좌표가 네번정도 변화가 없다면

#         액션 여러개인 부분 자르기



#     print("하나 끝")
#new_list.append(new_data)

#아직 final 작성x
```
