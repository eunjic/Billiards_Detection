```python
# connection
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017')


```

```python
#databas
db = client['Billiards']
```

```python
#collection
collection = db.all_positions
```

```python
#info먼저 넣기
import json
import os

output_path = "C:/Users/iai/Desktop/video_extraction/outputs"

for video in os.listdir(output_path):
    
    action_path = output_path + "/"+ video
    for action_name in os.listdir(action_path):
        
        action_folder = action_path + "/" + action_name
        #output 다 들어있는 폴더 진입
        
        with open(action_folder+'/info.json', encoding = 'utf-8') as json_file:
            info = json.load(json_file)
        
        with open(action_folder+'/'+action_name+ '.json', encoding = 'utf-8') as json_file:
            data = json.load(json_file)
            #list 형식

        #object 하나 insert
        #ObjectID는 임의로 저장됨
        new_position = {
            "_id" : video+action_name,
            "coordinate_set" : data,
            "skill" : info["skill"],
           
        } 
        #"score" : info["score"],
        #"kiss" : info["kiss"]   
        collection.insert_one(new_position)
```

```python
#단일 document/ object 찾기
import pprint
pprint.pprint(collection.find_one())
```
